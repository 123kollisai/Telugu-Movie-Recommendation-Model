```python
import pandas as pd
import numpy as np
import difflib

from sklearn.neighbors import NearestNeighbors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies=pd.read_csv("TeluguMovies_dataset (1).csv")
movies.head()
```

```python
movies=movies.drop("Rating",axis=1)
```

```python
movies.head()
```

```python
user_ratings=pd.read_csv("user_movie_ratings_40_per_user.csv")
user_ratings.head()
```

```python
user_ratings.rename(columns={"movie_id": "movieid"}, inplace=True)
user_ratings.head()
```

```python
movies=movies.merge(user_ratings,on='movieid')
movies.head()
```

```python
user_item_matrix = user_ratings.pivot(index=['user_id'], columns=['movieid'], values='rating').fillna(0)
user_item_matrix

```

```python

# Define a KNN model on cosine similarity

cf_knn_model= NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10, n_jobs=-1)


# Fitting the model on our matrix
cf_knn_model.fit(user_item_matrix)
```

```python
def get_closest_movie_id(movie_name, movies):
    highest_similarity = 0.0
    closest_movie_id = None
    
    for i, title in enumerate(movies['Movie']):
        similarity = difflib.SequenceMatcher(None, title.lower(), movie_name.lower()).ratio()
        if similarity > highest_similarity:
            highest_similarity = similarity
            closest_movie_id = movies['movieid'].iloc[i]
    
    if closest_movie_id is None:
        raise ValueError("No matching or similar movie found")
    
    return closest_movie_id

def movie_recommender_engine(movie_name, movies, cf_model, n_recs):
    # Create the interaction matrix
    matrix = movies.pivot_table(index='movieid', columns='user_id', values="rating").fillna(0)
    
    # Fit model on matrix
    cf_model.fit(matrix)
    
    # Extract input movie ID using simple string matching
    closest_movie_id = get_closest_movie_id(movie_name, movies[['movieid', 'Movie']].drop_duplicates())
    
    # Calculate neighbour distances
    distances, indices = cf_model.kneighbors(matrix.loc[[closest_movie_id]], n_neighbors=n_recs)
    movie_rec_ids = sorted(list(zip(indices.squeeze().tolist(), distances.squeeze().tolist())), key=lambda x: x[1])[::-1]
    
    # List to store recommendations
    cf_recs = []
    for idx, dist in movie_rec_ids:
        movie_title = movies.loc[movies['movieid'] == matrix.index[idx], 'Movie'].values[0]
        cf_recs.append({'Title': movie_title, 'Distance': dist})

    # Create DataFrame with recommendations
    df_recs = pd.DataFrame(cf_recs, index=range(1, n_recs + 1))
    
    return df_recs
```

```python
def get_user_movie_input():
    # Prompt the user to enter a movie name
    return input("Enter the name of the movie you like: ")

def movie_recommender_from_user_input(movies, cf_model, n_recs):
    # Get the movie name input from the user
    movie_name = get_user_movie_input()
    
    # Get the movie recommendations
    movie_recommendations = movie_recommender_engine(movie_name, movies, cf_model, n_recs)
    
    # Print the recommendations
    print(movie_recommendations[::-1])

# Example usage:
n_recs = 10
movie_recommender_from_user_input(movies, cf_knn_model, n_recs)
```
